```python
import pandas as pd
import numpy as np
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise import Reader
from surprise.model_selection import train_test_split
```

```python
ratings=pd.read_csv("C:\\Shakeer\\ML\\Assignment 3\\ml-latest-small/ratings.csv")
ratings.head()
```

# 1.	Create recommender system using ratings.csv file.

```python
reader=Reader(rating_scale=(1,5))
data=Dataset.load_from_df(ratings[['userId','movieId','rating']],reader)
data
```

```python
[trainset, testset]=train_test_split(data,test_size=15,shuffle=True)
```

```python
recom=KNNWithMeans(k=50,sim_options={'name':'cosine','user_based':True})
recom.fit(data.build_full_trainset()) 
test_pred=recom.test(testset)
RMSE=accuracy.rmse(test_pred)
```

# 2.	Display predicted rating for a particular userID and MovieID combinations (both taken as keyboard input)

```python
userid=int(input('Enter the user_id :'))
movieid=int(input('Enter the movie_id :'))
pre=recom.predict(userid,movieid)
print(pre)
print("The rating value for movieid", movieid, "by user", userid,"is", pre[3])
```

# 3.	Recommend a user to watch top 10 movies, which has not watched yet (as per his rating predictions). Take userID as a keyboard input. Fix the threshold rating as 2.5

```python
ratings.info()
```

```python
ratings_pivot=pd.pivot(ratings,index='userId',columns='movieId',values='rating').fillna(0)
```

```python
ratings_pivot.head()
```

```python
ratings_pivot.shape
```

```python
for i in range(1,610):
    RatedMovieId=list(ratings_pivot.loc[i,:][ratings_pivot.loc[i,:]!=0.0].reset_index()['movieId'])
    for j in RatedMovieId:
        ratings_pivot[j][i]=-1
```

```python
ratings_pivot
```

```python
ZeroRatedMovieId=list(ratings_pivot.loc[1,:][ratings_pivot.loc[1,:]==0.0].reset_index()['movieId'])
print(ZeroRatedMovieId[2])
print(len(ZeroRatedMovieId))
```

```python
for i in range(1,610):
    ZeroRatedMovieId=list(ratings_pivot.loc[i,:][ratings_pivot.loc[i,:]==0.0].reset_index()['movieId'])
    for j in ZeroRatedMovieId:
        pre=recom.predict(i,j)
        ratings_pivot[j][i]=pre[3]
        #print("The rating value for movieid",j ,"by user",i,"is", pre[3])
```

```python
ratings_pivot
```

# Top 10 movies with threshold rating as 2.5

```python
userid=int(input("Enter the userid: "))
TopRatedMovieratings=ratings_pivot.loc[userid,:][ratings_pivot.loc[userid,:]>=2.5].reset_index()
TopRatedMovieratings.rename(columns={userid:'rating'}).sort_values(by=['rating'],ascending=False).head(10)
```
